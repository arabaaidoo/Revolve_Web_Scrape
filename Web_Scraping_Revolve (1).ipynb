```python
# Grabbing all social links
social_links = []

for link in soup.find_all('a', href=True):
    href = link['href']
    if 'instagram.com' in href or 'facebook.com' in href or 'twitter.com' in href or 'pinterest.com' in href or 'tiktok.com' in href:
        social_links.append(href)


for social_link in social_links:
    print(social_link)
```

```python
# Find navigation bar categories
top_menu_links = []


for nav_item in soup.find_all('a'):
    nav_name = nav_item.get_text().strip()
    nav_link = nav_item.get('href')

    if nav_name and nav_link and len(nav_name) > 2:  # filtering empty links
        if nav_link.startswith('/'):
            top_menu_links.append((nav_name, "https://www.revolve.com" + nav_link))

print("Top Menu Links:")
for name, link in top_menu_links:
    print(f"{name}: {link}")
```
